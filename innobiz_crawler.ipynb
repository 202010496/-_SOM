```python
sales_df = pd.DataFrame()
product_df = pd.DataFrame()
classification_df = pd.DataFrame()
patent_df = pd.DataFrame()
certify_df = pd.DataFrame()
code_df = pd.DataFrame()
```

```python
money = []

for i in range(len(sales)):
    test = pd.DataFrame([sales[i].split()[7::7], sales[i].split()[8::7], sales[i].split()[9::7], sales[i].split()[10::7], sales[i].split()[11::7], sales[i].split()[12::7], sales[i].split()[13::7]])
    test.index = ['결산년도', '총자산', '자본금', '자본총계', '매출액', '영업이익', '당기순이익']
    money.append(test.transpose())
```

```python
sales_df['재무'] = money
product_df['상품'] = product
classification_df['산업군'] = classification
patent_df['특허'] = patent
certify_df['인증'] = certify
code_df['코드'] = code
```

```python
data4 = pd.concat([code_df, sales_df], axis=1)
data4 = pd.concat([data4, product_df], axis=1)
data4 = pd.concat([data4, classification_df], axis=1)
data4 = pd.concat([data4, patent_df], axis=1)
data4 = pd.concat([data4, certify_df], axis=1)
```
